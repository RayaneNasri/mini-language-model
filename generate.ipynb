```python
import torch

from src.models import Mlm_RM
from src.tokenizer import CharTokenizer
```

```python
# load text -- 
with open("data/shakespeare_preprocessed.txt", "r", encoding="utf-8") as f:
    text = f.read()

# defining the tokenizer --
tokenizer = CharTokenizer(text)
```

```python
# device 
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
```

```python
context_window= 64
embedding_dim= 32
hidden_size= 256
num_layers=2

model = Mlm_RM(
    model_type='lstm',
    vocab_size= len(tokenizer.vocabulary),
    embedding_dim=embedding_dim,
    hidden_size=hidden_size,
    num_layers=num_layers,
    device=device
)

model.load_state_dict(torch.load('parameters/rm_model.pth', map_location=device))
model.eval()
```

```python
model.generate('first ', tokenizer.char_to_int, tokenizer.int_to_char, max_length=200, temperature=1.)
```
